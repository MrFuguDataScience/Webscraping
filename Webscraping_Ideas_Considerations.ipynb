# `Anti-Scraping Tools:`

# `Headers and Importance for Webscraping:`

+ When the browser makes a connection (request) to the server, with the HTTP header.
    + Then the server will respond with relevant data with respect to the header type.

# `User-Agent:`

+ String containing the browser and operating system connecting to the server
+ When the browser makes a connection (request) to the server, with the HTTP header.
    + Then the server will respond with relevant data with respect to the header type.
+ Think of it like this, your saying "hello I am a Chrome browser or mobile version" where you are what can be thought of as a signature featuring information such as `device type, software version, operating system, application type`. 
    + Different browsers will use differing formats for each string representation
    
+ There is **something important to note** that different `user-agents` can show different information for the same website server this can be shown in some cases for mobile devices.

**Ex.)**

`Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36`

# `Headless Browser:`

Think of this as ignoring the front end GUI and working directly with the backend. You are essentially working via command line and trying to maximize your chances of data collection. 
+ This can be beneficial because you are trying to automate your data collection and avoid being caught. 

# `Why a Headless Browser?`

Let us consider what happens when we make a request to a webpage:
+ Usually we make a call to a webpage which is attached to a server with our HTML
+ The header is a piece of information sent to the server via HTML request. This can easily let the website know you are not a human.
    + A specific piece of information that is attached is the `User-Agent` and will describe who is making the request.
        + From this request the website can determine you are usnig `cURL` for example
      

When you use a request such as `cURL` you are using headers to retrieve data and sending what is essentially your signature we can say. Here is what it would look like on the backend.

**Ex.) To see what a Header Looks Like with User-Agent:** 

[To Check Your Header Response](http://www.httpbin.org/headers?json)

`------------------------------------------`

**`Pro's:`**
+ Easier to extract data
+ Load CSS or Java Script (for speed) and avoid HTML rendering (which may be your problem!).


**`Cons:`**
+ Simply getting blocked by the website which is a possibility (due to fingerprinting)
+ `Debugging`
+ `Front-end issues such as GUI`

# `NoneType Error Beautiful Soup:` 

+ This error has crept up in the past year, figuring why this keeps occuring is frustrating.
    + `When you see this error you will need to understand you are calling something UNAVAILABLE` this can be for a few reason.
    
    1.) You may be calling something hidden or blocked by the website that you may not know when webscraping
    
    2.) There may be a piece in the code that tricking you to think you are calling something when you actually cannot.
    
    3.) In some cases an exception will be needed to handle this because, you are accessing something that may be available but not for every instance when you are using `find_all`
