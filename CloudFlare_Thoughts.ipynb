# Cloudflare

*Handling* `Cloudflare` *Protection*:

`Cloudflare` uses various techniques to detect and block bots, such as JavaScript challenges and IP rate limiting.

*Basic understaning* of how `Cloudflare` operates: 

`--------------------------------------`

They are a cyber security and content delivery network. Because of this you are at a disposition as a webscraper because you are not spared from their wrath. Using `Selenium WebDriver` is a warning beacon making your presence known. `CloudFlare` uses *active* and *passive* detecting measures to counter any perceived threat.

*They utilize*:

+ *`Active` monitoring:* CAPTCHA, event tracking, (canvas) fingerprinting to name a few
+ *`Passive` monioring:* HTTP headers, TLS finger printing, checking your IP adress reputation, among many others
+ May imploy behavior techniques such as *mouse movements*
+ *JavaScript Challenges*


I want to mention a few important tasks that may occur when using this platform when you are using `automation or test software`

For example when you are trying to webscrape there are signatures that are looked for to determine your behavior and finger print.

+ When you use `Selenium` you have what is called a `WebDriver` that is used to connect a browser such as Chrome, Firefox, Mozilla, etc to that server through requests.
    + By default `Selenium` will send information to the website letting it know you are using active test-autiomation software. This trigger can directly get you into a hiccup with the server. Yes, you can mask this by turning off the ping but, it is not fool proof.

*Other behaviors can trigger this as well:*
    
+ When you try to connect to a server; information about you is given to them such as:
    + what version of browser, what type of browser you are on, screen resolution, mouse movements, timing of your responses like clicks and moving to new pages quickly, geo-location, IP address.
    
*Think of it like this*: we are creating an active digital finger print. This finger print consists of various markers that make the site able to detect who we are. Because of this you have to consider that sites can evaluate your IP Address as credible or potentially a bot (like a credit system) and they hold the keys to let you pass.

+ If you want to obscure your presence and fly under the radar it is imperative to include various steps in your process to roam around without raising suspicion.
